```python
board = list(range(1,10))

def draw_board(board):
    print('-' * 13)
    for i in range(3):
        print('|', board[0+i*3], '|', board[1+i*3],'|', board[2+i*3],'|')
        print('-' * 13)


def p_input(p_number):
    valid=False
    while not valid:
        p_ans = input('Выберете куда поставить ' + p_number + '?')
        try:
            p_ans = int(p_ans)
        except:
            print('Некорректный ввод')
            continue
        if p_ans >= 1 and p_ans <=9:
            if (str(board[p_ans-1]) not in 'XO'):
                board[p_ans-1] = p_number
                valid = True
            else:
                print('Эта клетка занята')
        else: 
            print('Некорректный ввод')
            

def win(board):
    win_c = ((0,1,2), (3,4,5), (6,7,8), (0,3,6), (1,4,7), (2,5,8), (0,4,8), (2,4,6))
    for each in win_c:
        if board[each[0]] == board[each[1]] == board[each[2]]:
            return board[each[0]]
    return False


def main(board):
    counter = 0
    won = False
    while not won:
        draw_board(board)
        if counter % 2 == 0:
            p_input('X')
        else:
            p_input('O')
        counter += 1
        if counter > 4:
            y = win(board)
            if y:
                print(y, 'выиграл')
                won = True
                break
        if counter == 9:
            print('Ничья')
            break
    draw_board(board)
main(board)
    
```
